# CHAPITER II: Machine Learning

## 1) Importing the necessary libraries

```python
from sklearn.preprocessing import RobustScaler, StandardScaler, MinMaxScaler, PolynomialFeatures, OneHotEncoder,LabelEncoder
from sklearn.feature_selection import RFECV
from sklearn.model_selection import train_test_split, GridSearchCV, learning_curve, RandomizedSearchCV
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.svm import LinearSVC, SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import VotingClassifier, AdaBoostClassifier, BaggingClassifier, StackingClassifier, RandomForestClassifier, GradientBoostingClassifier, ExtraTreesClassifier, HistGradientBoostingClassifier
from sklearn.metrics import f1_score, confusion_matrix, accuracy_score, recall_score, classification_report
```

## 2) Pre-processing

### Loading data

```python
df = data.copy()
bacup = data.copy()
df.shape
```

### Data cleaning

```python
df = df.drop(['N°', 'Nom et Prénoms', 'prof père', 'prof mère', 'Rang', 'RANG'], axis=1)
```

```python
df.shape
```

### Preprocessing

```python
# fill or drop a row with missing value
def imputation(data):
    return data.dropna(axis=0)

df = imputation(df)
df.shape
```

```python
# Categorical variable transform
"""def encoding(data, col):
    return data[col].astype("category").cat.codes"""
def encoding(data, categorical_columns):
    import joblib
    encoders = {}
    for col in categorical_columns:
        le = LabelEncoder()
        data[col] = le.fit_transform(data[col])
        encoders[col] = le 
        print(col)
    joblib.dump(encoders, 'encoders.pkl')
    print("encoder exported")
    return data
```

```python
def preprocessing(data, cat_var_list):
    encoding(data, cat_var_list)
    return data

# List of the categorical variables
cat_var_list = list(df.select_dtypes(include= 'object').columns)

# Data after transformation
data = preprocessing(df, cat_var_list)
data.head()


```

### Data split

```python
def data_split(data):
    # Data split
    trainset, testset = train_test_split(data , test_size=0.3, random_state=0)

    #Checking the proportion of classes in the target variable
    print(trainset.shape)
    print(testset.shape)
    print(trainset['Opt'].value_counts(normalize=True))
    print(testset['Opt'].value_counts(normalize=True))

    # Xtrain, ytrain, Xtest, ytest
    X_train = trainset.drop('Opt', axis=1)
    y_train = trainset['Opt']
    X_test = testset.drop('Opt', axis=1)
    y_test = testset['Opt']
    return X_train, y_train, X_test, y_test
    
```

```python
X_train, y_train, X_test, y_test =data_split(data)
```

### Modelling [using for tuner the data]

```python
# Modelization end evaluation function
def mod_evaluation(model):
    model.fit(X_train, y_train)
    y_pred=model.predict(X_test)
    print("Confusion matrix : ")
    print(confusion_matrix(y_test, y_pred))
    print("Classification_report : ")
    print(classification_report(y_test, y_pred, zero_division=0))
    N, train_score, val_score= learning_curve(model, X_train, y_train, train_sizes=np.linspace(0.2,1,10), scoring='accuracy', cv=5)
    plt.figure(figsize=(12,8))
    plt.plot(N, train_score.mean(axis=1), label='train score')
    plt.plot(N, val_score.mean(axis=1), label='validation score')
    plt.legend()
    plt.show()
```

```python
# Modelisation and evaluation
mod_evaluation(make_pipeline(StandardScaler(), SVC(random_state=0)))
```

### Feature selection

```python
# Feature selection (for importance of the variable in the estimator)
def selector(estimator):
    selector = RFECV(estimator, step = 1, min_features_to_select=9, cv=5)
    selector.fit(data.drop('Opt', axis=1), data['Opt'])
    return selector
```

```python

selector = selector(SGDClassifier(random_state=0))
print(selector.ranking_)
print(selector.support_)
```

```python
# Selectionned variable of this method
X = data.drop('Opt', axis=1)
X.columns[selector.support_]
```

```python
# Using data
data = pd.merge(data[X.columns[selector.support_]], data['Opt'], how='inner', on=data.index).drop('key_0', axis=1)
data
```

```python
# new data split
X_train, y_train, X_test, y_test =data_split(data)

# evaluation
mod_evaluation(make_pipeline(StandardScaler(), SVC(random_state=0)))
```

## 3) Modelling

### Creation of pipelines : model/feature engineering method/feature selection method/feature scaling 

```python
#preprocessor=make_pipeline(SelectFwe(f_classif, alpha=0.001))  #f_classif for the test ANOVA
linearSVC = make_pipeline(StandardScaler(), LinearSVC(dual=True ,multi_class='ovr', max_iter=10000))
log_reg = make_pipeline(StandardScaler(),LogisticRegression(solver='saga', multi_class='multinomial', max_iter=10000, C=0.01))
svm= make_pipeline(StandardScaler(),SVC(random_state=0))
KNeighbors= KNeighborsClassifier()
AdaBoost= AdaBoostClassifier(random_state=0)
#Stacking= StackingClassifier(estimators=100)
RandomForest= RandomForestClassifier(random_state=0, criterion='entropy')
GradientBoosting= GradientBoostingClassifier(random_state=0)
ExtraTrees= ExtraTreesClassifier(random_state=0)
#Voting= VotingClassifier([('class_SVC', class_SVC), ('KNeighbors', KNeighbors), ('RandomForest', RandomForest), ('DecisionTree', DecisionTree)], voting='soft')
Bagging= BaggingClassifier(n_estimators=100, random_state=0)
HistGradientBoost = HistGradientBoostingClassifier(random_state=0)
```

### List of all models used

```python

list_of_models= {
                'LinearSVC' : linearSVC,
                'LogisticRegression' : log_reg,
                'SVM': svm,
                'KNeighbors': KNeighbors,
                'AdaBoost': AdaBoost,
                'RandomForest': RandomForest,
                'GradientBoosting' : GradientBoosting,
                'ExtraTrees': ExtraTrees,
                'Bagging': Bagging,
                'HistGradientBoostingClassifier' : HistGradientBoost}
```

### Evaluation of all selected models with their learning curves

```python
for name, model in list_of_models.items():
    print(name)
    %time mod_evaluation(model)
```

## 4) Optimization of the model which has the best score after evaluation

#### Optimization of the model with GridSearchCV

```python
hyper_params= {'svc__C': [0.1, 1, 10],
              'svc__gamma': [1, 0.1, 0.01, 0.001],
              'svc__kernel': ['rbf','linear']}
model=RandomizedSearchCV(svm, hyper_params, cv=5, scoring= 'accuracy')
model.fit(X_train, y_train)
y_pred=model.predict(X_test)
print(" Best parameters :  {}".format(model.best_params_))
print("Classification report :")
print(classification_report(y_test, y_pred, zero_division=0))
```
